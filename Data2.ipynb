***Segunda Parte del Ejercicio***
En esta se procede a desarrollar la gráfica de la señal, su filtrado y el análisis requerido

```python
import urllib
from io import StringIO
from io import BytesIO
import csv
import numpy as np
from datetime import datetime
import matplotlib.pylab as plt
import pandas as pd
import scipy.signal as signal
```

```python
data=pd.read_csv('data.csv')
```

```python
data["Fecha"]=pd.to_datetime(data["Fecha"],format='%Y%m%d %H:%M:%S')
x=data['Fecha']
data.set_index(["Fecha"],inplace=True)
data.plot(figsize=(18,7))
plt.ylabel('Precio')
plt.grid()
plt.show()
```

***Procedimiento de Filtrado***
Comenzamos con el procedimiento de Filtrado de la Señal

Comenzamos con N=3, Wn=0.02

```python
N  = 3    # Orden del filtro
Wn = 0.02 # Corte de frecuancia
B, A = signal.butter(N, Wn)
```

```python
y=data['Transacciones']
```

```python
y=np.array(y)
y=y.astype(float)
```

```python
dat_filtrada=signal.filtfilt(B,A, y)
a=np.array(data.index)
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(211)
plt.plot(a,y, 'b-')
plt.plot(a,dat_filtrada, 'r-',linewidth=2)
plt.ylabel(r"Precios")
plt.legend(['Original','Filtrado'])
plt.title("Transacciones")
ax1.axes.get_xaxis().set_visible(False)
ax1 = fig.add_subplot(212)
plt.plot(a,y-dat_filtrada, 'b-')
plt.ylabel(r"Precios")
plt.xlabel("Fecha")
plt.legend(['Residuales'])
plt.show()
```

```python
plt.figure(figsize=(20,7))
ruido=y-dat_filtrada
corr=signal.correlate(ruido,ruido,mode="full")
plt.plot(corr[len(corr)//2:])
plt.show()
```
